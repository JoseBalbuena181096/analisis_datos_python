```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
mu = 10
sigma = 1

mu_2 = 15
sigma_2 = 3
```

```python
x = np.linspace(7, 13, 100) ## Crear un arreglo de 100 valores entre 7 y 13
x_2 = np.linspace(0, 20, 100) ## Crear un arreglo de 100 valores entre 10 y 20
```

```python
y = stats.norm.pdf(x, mu, sigma)
#y_2 = stats.norm.pdf(x_2, mu_2, sigma)
y_2 = stats.norm.pdf(x_2, mu, sigma_2)
```

```python
plt.plot(x,y, label='μ: 10,  σ:1')
plt.plot(x_2,y_2, label='μ: 15,  σ:1')

plt.xlim(2, 25)
plt.legend()
plt.show()
```
